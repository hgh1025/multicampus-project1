```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  # 한글 표시
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel("C:/mypycode/project/stat_150505.xls", skiprows=2) #파일 불러오기

#drop 불필요한 열을 삭제할수 있다.
df_slice = df.iloc[1: , [0,1,5,9,13,17,21,25,29,33,37,41,45,49,53,57,61,65,69,73,77]].drop([1,25,26], axis=0)  ## 전체 산업 자료 추출
df_slice

df=df_slice.rename({'Unnamed: 0':'직종'},axis=1)
df=df.set_index(['직종'])
df

# 컬럼 도출
data1 = df.columns[0:20]
list_data1 = list(data1)


#퍼센트 구하기
##df.transform(lamda x:100*x/x.sum())
df1 =df.div(df[list_data1].sum(axis=0),axis=1)*100
df1
```

```python
## 데이터
import pandas as pd
 
df = pd.DataFrame()
df['date'] = ['2004', '2005', '2006', '2007', '2008', '2009', '2010', 
              '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'] ## 날짜
df['sales'] =  ['2004', '2005', '2006', '2007', '2008', '2009', '2010', 
              '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'] ## 매출액
```

```python
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df['date'],df['sales']) ## 선그래프 생성
 
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('Sales for 10 days',fontsize=20) ## 타이틀 설정
plt.show()
```
