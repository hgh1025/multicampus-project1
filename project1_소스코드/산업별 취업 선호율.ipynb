```python
# 구직율
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np


matplotlib.rcParams['font.family']='Malgun Gothic'  # 한글 표시
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel("C:/mypycode/project/stat_150505.xls", skiprows=2) #skiprows 는 불필요한 행을 삭제.
df

# 데이터 전처리

df_slice = df.iloc[1: , [0,2,22,78]].drop([1,25,26] ,axis=0) # iloc[행 , 열] 추출 
df_slice  # 총계 행렬 삭제 , #drop 불필요한 열을 삭제할수 있다.


df=df_slice.rename({'Unnamed: 0':'직종', 'Unnamed: 2':'2002',  'Unnamed: 22':'2012', 'Unnamed: 78':'2021'},axis=1)
df=df.set_index(['직종']) #열을 index로 바꾸기
df


```

```python
# 컬럼 도출
data1 = df.columns[0:3]
list_data1 = list(data1)
list_data1 
```

```python
#퍼센트 구하기
##df.transform(lamda x:100*x/x.sum())
df=df.div(df[list_data1].sum(axis=0),axis=1)*100
df
```

```python
#그래프 구현

# 차트 종류, 제목, 차트 크기, 범례, 폰트 크기 설정
ax[0] = df.plot(kind='bar', title='산업별 구직 선호율', figsize=(12,4), legend=True, fontsize=12 ,grid= True)
ax[0].set_xlabel('산업', fontsize=12)          # x축 정보 표시
ax[0].set_ylabel('구직 선호율(%)', fontsize=12)     # y축 정보 표시
ax[0].legend(['2002년','2012년','2021년'], fontsize=12)    # 범례 지정

```

```python
#그래프 구현
# 차트 종류, 제목, 차트 크기, 범례, 폰트 크기 설정
ax = df.plot( title='산업별 구직 선호율', figsize=(12,4), legend=True, fontsize=12 ,grid= True)
ax.set_xlabel('산업', fontsize=12)          # x축 정보 표시
ax.set_ylabel('구직 선호율(%)', fontsize=12)     # y축 정보 표시
ax.legend(['2002년','2012년','2021년'], fontsize=12)    # 범례 지정

plt.show()
```
