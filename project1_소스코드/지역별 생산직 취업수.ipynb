```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['font.family']='Malgun Gothic'  # 한글 표시
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel("C:/mypycode/project/job_of_point.xlsx", skiprows=2) #파일 불러오기
df11 = df.drop([7],axis=0)
df11

```

```python

df1 = df11.set_index(['행정구역(시도)'])
df2 = df1.iloc[0:, [0,2,4,6,8]]
df2
```

```python

# 차트 종류, 제목, 차트 크기, 범례, 폰트 크기 설정
ax = df2.plot(kind='barh', title='지역별 생산직 취업수', figsize=(14,9), legend=True, fontsize=12 ,grid= True , width = 0.8)
ax.set_xlabel('취업수', fontsize=15)          # x축 정보 표시
ax.set_ylabel('행정구역(시도)', fontsize=15)     # y축 정보 표시

plt.show()
```
