```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['font.family']='Malgun Gothic'  # 한글 표시
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel("C:/mypycode/project/job_of_age1.xlsx", skiprows=2) #파일 불러오기
df
```

```python
df1 = df.iloc[[24,46,68,90] , [1,2,7,12,17]]
df1

```

```python
df1['연령']=['15-29세','30 - 39세','40 - 49세','50 - 59세']
df1
 
```

```python
df2=df1.rename({'총 계.1' : '직종'} ,axis=1 )
df3 = df2.set_index('연령')               
df3    
```

```python
#그래프 구현
# 차트 종류, 제목, 차트 크기, 범례, 폰트 크기 설정
ax = df3.plot(kind='bar', title='산업별 구인 비율', figsize=(15, 6), legend=True, fontsize=12)
ax.set_xlabel('연령', fontsize=12)          # x축 정보 표시
ax.set_ylabel('생산직 취업수', fontsize=12)     # y축 정보 표시
ax.legend(['2017년','2018년','2019년','2020년'], fontsize=12)    # 범례 지정

plt.show()
```

```python
position = [767,1258,1232,1099]
year = ['15-29세','30 - 39세','40 - 49세','50 - 59세']

fig = plt.figure(figsize=(8,8)) ## Figure 생성 
fig.set_facecolor('white')

colors = sns.color_palette('summer', len(year)) ## 바 차트 색상

xtick_label_position = list(range(len(year)))
plt.xticks(xtick_label_position, year)

plt.bar(xtick_label_position, position, color=colors)#bar차트

plt.plot(xtick_label_position, position, color='b', #plot 차트
         linestyle='--', marker='o') 

#그래프 구현
# 차트 종류, 제목, 차트 크기, 범례, 폰트 크기 설정
plt.xlabel('연령', fontsize=12)
plt.ylabel('생산직 취업수', fontsize=12)
plt.title('연령별 생산직 취업수', fontsize=20)
plt.legend(['2020년'], fontsize=12) # 범례 지정

plt.show()
```
